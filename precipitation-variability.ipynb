```python
import numpy as np
import os
import pandas as pd
filename = os.listdir('C:/Users/***/Desktop/500_100/all')
all_ = pd.DataFrame()
for name in filename:
    df = pd.read_csv('C:/Users/***/Desktop/500_100/all/{}'.format(name))
    all_[name] = df['prep']
df_1 = pd.read_csv('E:/data/all/18.251974_109.49696.csv')
df_1['time'] = pd.to_datetime(df_1['time']) 
df_1 = df_1.set_index('time')
mean = list(all_.mean(axis=1))
df_1['mean'] = mean
df_1 = df_1.drop(['lat', 'lon', 'prep'], axis=1)
years_1 = []
years_2 = []
years_3 = []
years_4 = []
years_5 = []
years_6 = []
for i in range(1961,1971):
    years_1.append(df_1['{}'.format(i)]['mean'].sum())
for i in range(1971,1981):
    years_2.append(df_1['{}'.format(i)]['mean'].sum())
for i in range(1981,1991):
    years_3.append(df_1['{}'.format(i)]['mean'].sum())
for i in range(1991,2001):
    years_4.append(df_1['{}'.format(i)]['mean'].sum())
for i in range(2001,2011):
    years_5.append(df_1['{}'.format(i)]['mean'].sum())
for i in range(2011,2020):
    years_6.append(df_1['{}'.format(i)]['mean'].sum())
results = []
sum_1 = 0
for i in range(10):
    sum_1 += abs(sum(years_1)/10-years_1[i])
sum_2 = 0
for i in range(10):
    sum_2 += abs(sum(years_1)/10-years_2[i])
sum_3 = 0
for i in range(10):
    sum_3 += abs(sum(years_1)/10-years_3[i])
sum_4 = 0
for i in range(10):
    sum_4 += abs(sum(years_1)/10-years_4[i])
sum_5 = 0
for i in range(10):
    sum_5 += abs(sum(years_1)/10-years_5[i])
sum_6 = 0
for i in range(9):
    sum_6 += abs(sum(years_1)/9-years_6[i])
results.append(sum_1/sum(years_1)*100)
results.append(sum_2/sum(years_2)*100)
results.append(sum_3/sum(years_3)*100)
results.append(sum_4/sum(years_4)*100)
results.append(sum_5/sum(years_5)*100)
results.append(sum_6/sum(years_6)*100)
df_results = pd.DataFrame(index = ['1961-1970','1971-1980','1981-1990','1991-2000','2001-2010','2011-2019'], 
                          columns = ['All of China','DongBei','HuaBei','HuaDong','HuaNan','HuaZhong','XiBei','XiNan'])
df_results['All of China'] = results
df_results.to_csv('precipitation variability.csv')
```
